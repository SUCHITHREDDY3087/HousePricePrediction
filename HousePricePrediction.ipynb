```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading data set
from google.colab import files
import io

print("Please upload the 'train.csv' file.")
uploaded = files.upload()

# Assuming 'train.csv' is the file you want to read
# If you uploaded multiple files, you might need to select the correct one.
if 'train.csv' in uploaded:
    data = pd.read_csv(io.BytesIO(uploaded['train.csv']))
    print("File 'train.csv' uploaded and loaded successfully.")
else:
    print("Error: 'train.csv' not found among uploaded files. Please ensure you upload the correct file with this name.")
```

```python
#used for understanding data structure
data.head()
```

```python
#it is used for getting basic information of data structure
data.info()
#it is used for getting statistical summary of numerical columns
data.describe()
```

```python
#selecting important features from the data set
features = ['LotArea','OverallQual','YearBuilt','TotalBsmtSF','GrLivArea','BedroomAbvGr','FullBath']

X = data[features]
y = data['SalePrice']
```

```python
#to handling the missing values int data
X=X.fillna(X.mean())
```

```python
#training and test spliting the model
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, random_state=50)
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
#predicting the model
predictions = model.predict(X_test)
```

```python
#model evaluation to get good score
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

print("Mean Squared Error:", mse)
print("R2 Score:", r2)
```

```python
#visualizing of model
plt.scatter(y_test, predictions)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs. Predicted Prices")
plt.show()
```

```python
#for saving the model for further use
import pickle
pickle.dump(model,open('model.pkl','wb'))
```
